```python
from researchplot import bar, line, scatter

data = [10, 20, 30, 40, 50]
bar(data, xlabel='Categories', ylabel='Values', format='nature', output_path='figure.png')
```

```python
from researchplot import loss_vs_epoch
import numpy as np

epochs = range(1, 11)
loss = np.random.rand(10)
val_loss = np.random.rand(10)
loss_vs_epoch(epochs, loss, val_loss=val_loss, format='nature', output_path='loss_vs_epoch.png')
```
